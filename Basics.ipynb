# Visualizing The Gender Gap In College Degrees

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(15, 24))

# comparing across all degrees using one 'for loop'
stem_counter = 0
lib_arts_counter = 0
other_counter = 0

for sp in range(0,18):
    if sp == 16:
        pass
    else:
        ax = fig.add_subplot(6,3,sp+1)
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom="off")
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        if sp in [0, 3, 6, 9, 12, 15]:
            ax.plot(women_degrees['Year'], 
                    women_degrees[stem_cats[stem_counter]], 
                    c=cb_dark_blue, 
                    label='Women', 
                    linewidth=3
                   )
            ax.plot(women_degrees['Year'], 
                    100-women_degrees[stem_cats[stem_counter]], 
                    c=cb_orange, 
                    label='Men', 
                    linewidth=3
                   )
            ax.set_title(stem_cats[stem_counter])
            ax.set_yticks([0, 100])
            ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
            if stem_counter == 0:
                ax.text(2005, 14, 'Men')
                ax.text(2004, 83, 'Women')
            elif stem_counter == 5:
                ax.text(2005, 87, 'Men')
                ax.text(2004, 8, 'Women')
                ax.tick_params(labelbottom="on")
            stem_counter += 1
        elif sp in [1, 4, 7, 10, 13]:
            ax.plot(women_degrees['Year'], 
                    women_degrees[lib_arts_cats[lib_arts_counter]], 
                    c=cb_dark_blue, 
                    label='Women', 
                    linewidth=3
                   )
            ax.plot(women_degrees['Year'], 
                    100-women_degrees[lib_arts_cats[lib_arts_counter]], 
                    c=cb_orange, 
                    label='Men', 
                    linewidth=3
                   )
            ax.set_title(lib_arts_cats[lib_arts_counter])
            ax.set_yticks([0, 100])
            ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
            if lib_arts_counter == 0:
                ax.text(2005, 21, 'Men')
                ax.text(2004, 76, 'Women')
            elif lib_arts_counter == 4:
                ax.tick_params(labelbottom="on")
            lib_arts_counter += 1
        elif sp in [2, 5, 8, 11, 14, 17]:
            ax.plot(women_degrees['Year'], 
                    women_degrees[other_cats[other_counter]], 
                    c=cb_dark_blue, 
                    label='Women', 
                    linewidth=3
                   )
            ax.plot(women_degrees['Year'], 
                    100-women_degrees[other_cats[other_counter]], 
                    c=cb_orange, 
                    label='Men', 
                    linewidth=3
                   )
            ax.set_title(other_cats[other_counter])
            ax.set_yticks([0, 100])
            ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
            if other_counter == 0:
                ax.text(2005, 6, 'Men')
                ax.text(2004, 92, 'Women')
            elif other_counter == 5:
                ax.text(2005, 62, 'Men')
                ax.text(2004, 32, 'Women')
                ax.tick_params(labelbottom="on")
            other_counter += 1
    
plt.savefig('gender_degrees.png')
plt.show()
```
